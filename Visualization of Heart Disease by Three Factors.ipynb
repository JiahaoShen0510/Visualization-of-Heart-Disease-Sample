```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data0=pd.read_csv("heart-disease.csv")
data0
```

```python
data0.describe()
```

```python
over_50=data0[data0["age"]>50]
```

```python
over_50.describe()
```

```python
over_50.sort_values("age")
```

```python
figure, (ax1,ax2,ax3)=plt.subplots(3,1,sharex=True,figsize=(10,10))

scatter1=ax1.scatter(over_50["age"],over_50["trestbps"],c=over_50["target"],cmap="binary")
ax1.set(ylabel="Resting Blood Pressure")

scatter2=ax2.scatter(over_50["age"],over_50["thalach"],c=over_50["target"],cmap="binary")
ax2.set(ylabel="Maximum Heart Rate ")

scatter3=ax3.scatter(over_50["age"],over_50["chol"],c=over_50["target"],cmap="binary")
ax3.set(xlabel="Age",ylabel="Cholesterol Test")

ax1.axhline(over_50["trestbps"].mean(),linestyle="--")
ax2.axhline(over_50["thalach"].mean(),linestyle="--")
ax3.axhline(over_50["chol"].mean(),linestyle="--")


ax1.legend(*scatter1.legend_elements(),title="target")

figure.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold");
```

```python
figure.savefig("Heart Disease Analysis.png")
```
